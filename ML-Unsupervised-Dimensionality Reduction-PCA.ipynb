```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the Penguins dataset
penguins = sns.load_dataset('penguins')

# Display the first few rows
print(penguins.head())

# Drop rows with missing values and select numerical features
penguins = penguins.dropna()
features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
X = penguins[features]

# Standardizing the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

# Applying PCA
pca = PCA(n_components=None)
pca_df = pca.fit(scaled_data)

pca_df.explained_variance_ratio_

pca = PCA(n_components=2)  # Reduce to 2 dimensions
pca_result = pca.fit_transform(scaled_data)

# Creating a DataFrame for PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df['species'] = penguins['species'].values

# Explained variance ratio
explained_variance = pca.explained_variance_ratio_

# Plotting the PCA result
plt.figure(figsize=(10, 7))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='species', palette='viridis', s=100)
plt.title('PCA of Penguins Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid()
plt.legend()
plt.show()

# Print explained variance
print('Explained variance by each principal component:', explained_variance)
```
